```python
# 15.Plot the distribution of purchase amounts using a histogram
import matplotlib.pyplot as plt
plt.hist(df['Amount'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Purchase Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 16.Plot total sales by Age Group (Bar Chart)
df.groupby('Age Group')['Amount'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 17.Pie chart of gender-wise sales contribution
df.groupby('Gender')['Amount'].sum().plot.pie(autopct='%1.1f%%')
plt.title('Gender-wise Sales Contribution')
plt.ylabel('')
plt.show()
```

```python
# 18.Histogram of Purchase Amounts
plt.hist(df['Amount'], bins=30, color='green')
plt.title('Purchase Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
```

```python
# 19.Countplot of Age Groups
sns.countplot(data=df, x='Age Group', palette='Set2')
plt.title('Customer Count by Age Group')
plt.show()
```

```python
# 20.Barplot of average sales by gender
sns.barplot(data=df, x='Gender', y='Amount', estimator='mean', palette='pastel')
plt.title('Average Sales by Gender')
plt.show()
```

```python
# 21.Boxplot of Amount by Marital Status
sns.boxplot(data=df, x='Marital_Status', y='Amount')
plt.title('Spending by Marital Status')
plt.show()
```

```python
# 22.Violin plot of Amount by Gender
sns.violinplot(data=df, x='Gender', y='Amount', palette='muted')
plt.title('Purchase Distribution by Gender')
plt.show()
```

```python
# 23.Line plot of top 5 states by revenue
top_states = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(5)
top_states.plot(kind='line', marker='o')
plt.title('Top 5 States by Revenue')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
# 24.Barh plot of revenue by zone
df.groupby('Zone')['Amount'].sum().sort_values().plot(kind='barh', color='coral')
plt.title('Revenue by Zone')
plt.xlabel('Amount')
plt.ylabel('Zone')
plt.show()
```

```python
# 25.Heatmap of correlation matrix (numeric columns)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 26.Scatterplot of Age vs Amount (if numeric Age is present)
sns.scatterplot(data=df, x='Age', y='Amount', hue='Gender')
plt.title('Age vs Purchase Amount')
plt.show()
```

```python
# 27.Distribution plot of Amount column
sns.histplot(df['Amount'], bins=30, kde=True)
plt.title('Amount Distribution')
plt.show()
```

```python
#28.Compare sales by Gender using groupby bar plot
df.groupby('Gender')['Amount'].sum().plot(kind='bar', color=['pink', 'blue'])
plt.title('Gender-wise Total Sales')
plt.show()
```

```python
#29.Plot top 5 product categories by sales
top_products = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='bar', color='orange')
plt.title('Top 5 Product Categories by Sales')
plt.show()
```

```python
# 30.Boxplot of Amount vs Age Group
sns.boxplot(data=df, x='Age Group', y='Amount')
plt.title('Amount by Age Group')
plt.show()
```

**KEY FINDINGS**

✅ Key Findings from Diwali Sales Data
Most Revenue Comes from 26–35 Age Group
This group contributed the highest total sales, indicating they are the prime target audience.

Males and Married Customers Spend More
On average and in total, male and married customers outspent female and unmarried ones.

High Variation in Purchase Amounts
NumPy analysis shows high standard deviation and outliers. Most purchases are between ₹500–₹5000, but some large transactions skew the average upward.

Top States Drive Majority of Sales
A few states like Maharashtra, Delhi, and Karnataka contribute the most revenue, while others have low engagement.

Product Sales are Unevenly Distributed
Electronics and Clothing categories dominate; others like Books or Toys have minimal revenue.

Purchase Behavior is Not Strongly Age-Correlated
Scatter plots and boxplots show people across age groups spend both low and high amounts — no clear linear trend with age.

Visualizations Confirm Skewed Spending Patterns
Histograms and KDE plots confirm that most users make smaller purchases; big spenders are fewer but impactful.


🎯 **CONCLUSION**

The Diwali sales data paints a clear picture of modern consumer behavior: young, married male shoppers aged 26–35 are the driving force behind revenue. While the majority of purchases are modest in value, a handful of high-ticket transactions significantly elevate the overall sales figures — revealing a classic case of the 80/20 rule in action.

Geographically, sales are concentrated in a few high-performing states, suggesting untapped market potential in other regions. Product-wise, categories like Electronics and Clothing reign supreme, while others remain underexplored.

This analysis not only highlights who buys and how much, but also where brands can focus their marketing, diversify product strategy, and target growth opportunities in future campaigns.
